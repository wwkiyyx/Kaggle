# opencv 读取 matplotlib 显示图像   
 - opencv 是 bgr 正常显示是 rgb

```python
import cv2
from matplotlib import pyplot as plt

print(cv2.__version__)

def showGray(mat) :
    mat_color = cv2.cvtColor(mat, cv2.COLOR_GRAY2RGB)
    mat_show = mat_color[:,:,::-1]
    plt.imshow(mat_show)

def showColor(mat) :
    mat_show = mat[:,:,::-1]
    plt.imshow(mat_show)
    
color = cv2.imread('../input/wangwenkai40/CocaCola/1_YES/IMG_20210615_193042_713.jpg', cv2.IMREAD_COLOR)
showColor(color)
```

```python
display(color)
```
